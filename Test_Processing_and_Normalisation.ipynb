```python
df = mail_df[["text"]]
df["text"] = df["text"].astype(str)
df.head()
```

## Upper casing & Lower casing

```python
df["upper_text"] = df["text"].str.upper()
df.head()
```

```python
df["lower_text"] = df["text"].str.lower()
df.head()
```

## Punctuation Removal

```python
import string
```

```python
Punctuation = string.punctuation
Punctuation
```

```python
def remove_punctuation(text):
  return text.translate(str.maketrans("", "", Punctuation))
```

```python
df["text_punc_removed"] = df["lower_text"].apply(lambda text: remove_punctuation(text))
df.head()
```

## Stop words removal

```python
import nltk
nltk.download("stopwords")
from nltk.corpus import stopwords
```

```python
stopwrds = set(stopwords.words("english"))
stopwrds
```

```python
def remove_stopword(text):
  return " ".join([word for word in str(text).split() if word not in stopwrds])
```

```python
df["text_stopword_removed"] = df["text_punc_removed"].apply(lambda x: remove_stopword(x))
df.head()
```

## Freequent Words Removal

```python
from collections import Counter
cnt = Counter()
```

```python
for text in df["text_stopword_removed"].values:
  for word in text.split(): # Split each text into words
    cnt[word] += 1

cnt.most_common(10) # Get the 10 most common words and their counts
```

```python
df["text_stopword_removed"].values
```

```python
df["text_stopword_removed"]
```

```python
Freq = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    return " ".join([word for word in str(text).split() if word not in Freq])
```

```python
df["text_freq_removed"] = df["text_stopword_removed"].apply(lambda text: remove_freqwords(text))
df.head()
```

## Stemming

```python
from nltk.stem.porter import PorterStemmer

Pstemmer = PorterStemmer()
```

```python
def po_stemmer(text):
  return " ".join([Pstemmer.stem(word) for word in text.split()])
```

```python
df["text_po_stem"]= df["text_freq_removed"].apply(lambda text: po_stemmer(text))
df.head(1)
```

```python
from nltk.stem.snowball import SnowballStemmer

Sstemmer = SnowballStemmer("english")
```

```python
def stem_words(text):
    return " ".join([Sstemmer.stem(word) for word in text.split()])

df["text_stemmed"] = df["text_freq_removed"].apply(lambda text: stem_words(text))
df.head(2)
```

## Tokenization

```python
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
def tokenize_words(text):
    return word_tokenize(text)

df["tokenized_default"] = df["text"].apply(lambda text: tokenize_words(text))
df["tokenized_default"].head()
```

```python
from nltk.tokenize.toktok import ToktokTokenizer
tokenizer_tok = ToktokTokenizer()
```

```python
def tokenize_words(text):
    return tokenizer_tok.tokenize(text)

df["tokenized_tok"] = df["text"].apply(lambda text: tokenize_words(text))

df["tokenized_tok"].head()
```

## Chat Word Corrction

```python
chat_words_str = """
AFAIK=As Far As I Know
AFK=Away From Keyboard
ASAP=As Soon As Possible
ATK=At The Keyboard
ATM=At The Moment
A3=Anytime, Anywhere, Anyplace
BAK=Back At Keyboard
BBL=Be Back Later
BBS=Be Back Soon
BFN=Bye For Now
B4N=Bye For Now
BRB=Be Right Back
BRT=Be Right There
BTW=By The Way
B4=Before
B4N=Bye For Now
CU=See You
CUL8R=See You Later
CYA=See You9
FAQ=Frequently Asked Questions
FC=Fingers Crossed
FWIW=For What It's Worth
FYI=For Your Information
GAL=Get A Life
GG=Good Game
GN=Good Night
GMTA=Great Minds Think Alike"""
```

```python
chat_words_map_dict = {}
chat_words_list = []
for line in chat_words_str.split("\n"):
    if line != "":
        cw = line.split("=")[0]
        cw_expanded = line.split("=")[1]
        chat_words_list.append(cw)
        chat_words_map_dict[cw] = cw_expanded
chat_words_list = set(chat_words_list)

def chat_words_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words_list:
            new_text.append(chat_words_map_dict[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

## Spelling Correction

```python
!pip install pyspellchecker
```

```python
from spellchecker import SpellChecker

spell = SpellChecker()
```

```python
def correct_spellings(text):
    corrected_text = []
    misspelled_words = spell.unknown(text.split())
    for word in text.split():
        if word in misspelled_words:
            corrected_text.append(spell.correction(word))
        else:
            corrected_text.append(word)
    return " ".join(corrected_text)
```

```python
text = "speling correctin"
correct_spellings(text)
```

## Removal Of URL

```python
import re
```

```python
def remove_urls(text):
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    return url_pattern.sub(r'', text)
```
